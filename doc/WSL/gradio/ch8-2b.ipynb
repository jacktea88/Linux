```python
import gradio as gr

def greet(name, is_lady, fahrenheit):
    if is_lady:
        greeting = name + "女士你好!"
    else:
        greeting = name + "男士你好!"
    greeting += " 今天溫度是華氏: " + str(fahrenheit)
    celsius = (fahrenheit - 32) * 5 / 9
    return greeting, round(celsius, 2)

app = gr.Interface(fn=greet, 
         inputs=["text", "checkbox", gr.Slider(0, 100)],
         outputs=["text", "number"])
app.launch()
```
